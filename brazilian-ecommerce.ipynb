```python
# 5. Manage tables with null values (Orders, Reviews, and Products tables)

# 5.1 Orders table

print("=== Null Audit===")
null_report = pd.DataFrame({
    'null_count' : orders.isnull().sum(),
    'null_percentage' : (orders.isnull().sum()/len(orders)*100).round(2)
})
print(null_report[null_report['null_count']>0])

## Why are they null?
print("\n=== WHY ARE THEY NULL? ===")
print(orders['order_status'].value_counts())
## delivered_date is null because the order is cancelled/shipped/processing
## That is VALID data — not an error
```

```python
## Double check whether delivered orders have null delivery date.
invalid_delivered = orders[
    (orders['order_status'] == 'delivered') &
    (orders['order_delivered_customer_date'].isna())
]

print("Invalid delivered rows:", len(invalid_delivered))
```

```python
## Look up for the 8 invalid delivered values and why they're null.
invalid_delivered = orders[
    (orders['order_status'] == 'delivered') &
    (orders['order_delivered_customer_date'].isna())
]

invalid_delivered
```

```python
## Flag invalid delivered orders (status = delivered but no delivery date)
## 8 rows out of 99k — kept in dataset, excluded in delivery analysis.
orders['is_invalid_delivered'] = (
    (orders['order_status'] == 'delivered') &
    (orders['order_delivered_customer_date'].isna())
)
```

```python
# 5.2 Reviews table

print("=== Null Audit===")
null_report = pd.DataFrame({
    'null_count' : reviews.isnull().sum(),
    'null_percentage' : (reviews.isnull().sum()/len(reviews)*100).round(2)
})
print(null_report[null_report['null_count']>0])

## Why are they null?, read it as story.
print("\n=== WHY ARE THEY NULL? ===")
print(reviews['review_score'].value_counts())
## reviews title and message are null because the customers mostly prefer to leave just a quick score rather than a whole message review
## That is VALID data — not an error.
```

```python
## Allies commented reviews for further analysis.
reviews['has_comment'] = reviews['review_comment_message'].notna()

## Explore whether the low-rated reviews are more commented rather than high-rated and vice versa, for further understanding.
reviews.groupby('has_comment')['review_score'].mean()

reviews.groupby('has_comment')['review_score'].value_counts(normalize=True)
```

```python
## Fill null values.
reviews['review_comment_title'] = reviews['review_comment_title'].fillna('')
reviews['review_comment_message'] = reviews['review_comment_message'].fillna('')
```

```python
# 5.3 Products table

print("=== Null Audit===")
null_report = pd.DataFrame({
    'null_count' : products.isnull().sum(),
    'null_percentage' : (products.isnull().sum()/len(products)*100).round(2)
})
print(null_report[null_report['null_count']>0])

## Why are they null?
## Some products have missed their characteristics, which can be considered an error.
```

```python
## Fill missing category names with 'unknown' — 610 rows, needed for analysis.
products['product_category_name'] = products['product_category_name'].fillna('unknown')

## Drop 2 rows with missing physical measurements — cannot be estimated, minor loss.
products.dropna(subset=['product_weight_g'], inplace=True)
```

```python
# 6. Collapse payments to order level

## Save original payments before we aggregate, so we can still grab payment_type later.
payments_original = payments.copy()

## Sum payment_value per order. Orders can have multiple payments (voucher + card etc.).
payments = payments.groupby('order_id', as_index=False)['payment_value'].sum()

## Keep the payment_type of the biggest payment as the order's main method.
payments_types = payments_original.sort_values('payment_value', ascending=False)\
    .drop_duplicates(subset='order_id', keep='first')[['order_id', 'payment_type']]

## Attach payment_type back to the summed payments.
payments = payments.merge(payments_types, on='order_id')

## Check: one row per order after aggregation.
assert payments['order_id'].is_unique, "Duplicate order_ids still present"
```

```python
product_category.loc[len(product_category)] = ['unknown', 'unknown']
```

```python
# 7. Merge all tables into one flat dataset for quick exploratory queries
olist_df = (
    orders
    .merge(customers[['customer_id','customer_city', 'customer_state']], on='customer_id', how='left')
    .merge(order_items[['order_id','product_id','seller_id','price', 'freight_value']], on='order_id', how='left')
    .merge(payments[['order_id','payment_type', 'payment_value']], on='order_id', how='left')
    .merge(reviews[['order_id','review_score','review_comment_title','review_comment_message']], on='order_id', how='left')
    .merge(products[['product_id', 'product_category_name']], on='product_id', how='left')
    .merge(product_category, on='product_category_name', how='left')
)
olist_df.rename(columns={'product_category_name_english': 'product_category_en'}, inplace=True)

print(f'Orders original: {len(orders)}')
print(f'Olist merged shape: {olist_df.shape}')
olist_df.head()
```

```python
# 8. Load data into PostgreSQL for SQL analysis

from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:1212@localhost:5432/brazilian-ecommerce')

## Load flat dataset 
olist_df.to_sql('olist', engine, if_exists='replace', index=False)

## Load normalized tables into PostgreSQL for SQL analysis.
## Loading each CSV as its own table preserves relational structure,
## so SQL joins work properly and payment_value isn't duplicated across order_items rows.
orders.to_sql('orders', engine, if_exists='replace', index=False)
customers.to_sql('customers', engine, if_exists='replace', index=False)
order_items.to_sql('order_items', engine, if_exists='replace', index=False)
payments.to_sql('payments', engine, if_exists='replace', index=False)
reviews.to_sql('reviews', engine, if_exists='replace', index=False)
products.to_sql('products', engine, if_exists='replace', index=False)
sellers.to_sql('sellers', engine, if_exists='replace', index=False)
product_category.to_sql('product_category', engine, if_exists='replace', index=False)
geolocation.to_sql('geolocation', engine, if_exists='replace', index=False)


print('Done')
```
